The Script implements three parts of a personal savings simulator:
    Part A: Basic savings to down payment (no raise)
    Part B: Savings with a semi-annual salary raise
    Part C: Bisection search for optional savings rate (target: 36 months)

---
SHARED CONSTANTS (as used across all parts) include:

ANNUAL_RETURN = 0.04 # 4% annual investment returns
MONTHLY_RETURN = ANNUAL_RETURN / 12 # r/12
DOWN_PAYMENT_FRAC = 0.10 # 25% down payment required
SEMI_ANNUAL_RAISE-C = 0.07 # 7% raise - fixed for Part C
HOUSE_COST_C = 350,000 # 350k starter home (Galway, Ireland)
TARGET_MONTHS = 36 # 3-year savings goal - Part C
TOLERANCE = 100 # +- 100 euro accuracy for bisection

The follownig function prompts the user for a validated float value.

    min_value: float -> minimum acceptable value and default is 0.0
    max_value: float -> maximum acceptable value (optional) # If provided

```python
def get_float_input(prompt: str, min_value: float = 0.0, max_value: float = None) -> float:
    while True:
        try:
            value = float(input(prompt))

            if value < min_value:
                print(f"Value must be at least {min_value}. Please try again.")
                continue

            if max_value is not None and value > max_value:
                print(f"Value must be at most {max_value}. Please try again")
                continue

            return value

        except ValueError:
            print("Invalid input. Please enter a valid numeric value.")
```

With the nxt Fn, monthly savings accumulation is simulated until the deposit is reached.

For example:

Monthly recurrence ( 10% deposit target):
    S_{t+1} = S_t + (r/12) * S_t + (salary/12) * portion_saved

Salary upfate every 6 months (Relevant to Part B only):
    salary_{t+6} = salary_t * (1 + semi_annual_raise)

Key Parameters:

annual_salary -> This refers to gross annual salary,
portion_saved -> decimal fraction of mntly salary saved,
total_cost -> full property purchase price, and
semi_annual_raise -> signals the float raise every 6 mnts

Currency remains the Euro (€)

```python
def simulate_savings(
    annual_salary: float,
    portion_saved: float,
    total_cost: float,
    semi_annual_raise: float = 0.0
) -> int: # returns integer in num of mnts required to reach the deposit

    deposit = DOWN_PAYMENT_FRAC * total_cost
    current_savings = 0.0
    months = 0
    salary = annual_salary

    while current_savings < deposit:
        investment_return = current_savings * MONTHLY_RETURN
        monthly_contribution = (salary / 12) * portion_saved
        current_savings += investment_return + monthly_contribution
        months += 1

        if semi_annual_raise > 0 and months % 6 == 0:
            salary *= (1 + semi_annual_raise)

    return months
```
